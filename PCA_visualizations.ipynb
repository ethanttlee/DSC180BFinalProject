```r
# COLOC results
diseases_df <- read.csv("diseases_df")
diseases_df
```

```r
# Transform data
rownames(diseases_df) <- diseases_df$gene_name
diseases_df_numeric <- diseases_df[,-1]  # Remove the 'gene_name' column
diseases_df_numeric <- data.frame(lapply(diseases_df_numeric, as.numeric))
diseases_df_numeric
```

```r
# PCA
df_standardized <- scale(diseases_df_numeric)
pca_result <- prcomp(df_standardized)
summary(pca_result)
```

```r
# Scree Plot
plot(pca_result$sdev^2 / sum(pca_result$sdev^2), type='b', xlab="Principal Component", ylab="Variance Explained", main="Scree Plot")
```

```r
# Cumulative scree plot
# Component variance
var_explained <- pca_result$sdev^2 / sum(pca_result$sdev^2)

# Cumilative variance
cum_var_explained <- cumsum(var_explained)

plot(cum_var_explained, type='b', 
     xlab="Principal Component", 
     ylab="Cumulative Variance Explained", 
     main="Cumulative Scree Plot",
     ylim=c(0,1))
abline(h=0.8, col="red", lty=2)
```

```r
# biplot
biplot(pca_result, scale=0)
```

```r
# Create 3 clusters for visualization
data_for_clustering <- diseases_df[, -1]
distance_matrix <- dist(data_for_clustering)
hclust_result <- hclust(distance_matrix)

clusters_vector <- cutree(hclust_result, k=3)
clusters_vector
```

```r
# 2D PCA plot with clusters highlighted
colors <- c("#40E0D0", "#1E90FF", "#9932CC")
cluster_colors <- colors[clusters_vector]

plot(pca_result$x[, 1], pca_result$x[, 2], 
     xlab = "Principal Component 1", ylab = "Principal Component 2", 
     main = "Scatter Plot of the First Two Principal Components",
     col = cluster_colors, 
     pch = 19) 

legend("topright", legend = paste("Cluster", unique(clusters_vector)), col = colors, pch = 19)
```

```r
rownames(diseases_df) <- diseases_df$gene_name
diseases_df$Cluster <- clusters_vector

library(dplyr)
cluster_means <- diseases_df %>%
  group_by(Cluster) %>%
  summarise(across(everything(), mean, na.rm = TRUE))

# Mean PP_H4 values for each disease by cluster
cluster_means <- cluster_means[, !(names(cluster_means) %in% c("gene_name"))]
print(cluster_means)
```

```r
# Processing for plotting
cluster_means_long <- data.frame(Cluster = integer(), Disease = character(), Mean_PP_H4 = numeric())

for(disease in names(cluster_means)[-1]) { 
  temp_df <- data.frame(Cluster = cluster_means$Cluster,
                        Disease = disease,
                        Mean_PP_H4 = cluster_means[[disease]])
  
  cluster_means_long <- rbind(cluster_means_long, temp_df)
}

cluster_means_long$Cluster <- as.factor(cluster_means_long$Cluster)
cluster_means_long
```

```r
# Plot heatmap
library(ggplot2)

colors <- c("blue", "cyan3", "darkgreen") 

ggplot(cluster_means_long, aes(x = Disease, y = Cluster, fill = Mean_PP_H4)) +
  geom_tile() + 
  geom_text(aes(label = sprintf("%.2f", Mean_PP_H4)), color = "white", size = 3) + 
  scale_fill_gradientn(colors = colors) + 
  theme_minimal() +
  labs(title = "Mean PP_H4 Values by Cluster and Disease", y = "Cluster", x = "Disease") +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```
