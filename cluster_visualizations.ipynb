```r
diseases_df <- read.csv("diseases_df", header = TRUE)

# Hierarchichal Clustering
data_for_clustering <- diseases_df[, -1]
distance_matrix <- dist(data_for_clustering)
hclust_result <- hclust(distance_matrix)

# Dendrogram
plot(hclust_result, hang = -1, cex = 0.6)
par(cex.axis = 0.0001)
```

```r
# Save the dendrogram as a PNG file with transparent background
dendogram <- plot(hclust_result, hang = -1, cex = 0.6)

png("dendogram.png", width = 800, height = 600, bg = "transparent")
print(dendogram)
dev.off() 
```

```r
plot(hclust_result, hang = -1, cex = 0.6, show_columnnames = FALSE)
```

```r
# Hierarchical clustering
hclust_result <- hclust(distance_matrix)

# 3 clusters
clusters <- cutree(hclust_result, k = 3)
print(clusters)
```

```r
# Heatmap
data_for_heatmap <- diseases_df[, -1]
my_palette <- colorRampPalette(c("#004c6d", "#2a9d8f", "#f4a261", "#ee6c4d", "#9d0208"))(100)
pheatmap(data_for_heatmap, cluster_rows = TRUE, cluster_cols = TRUE, show_rownames = FALSE, color = my_palette)
```

```r
heatmap <- pheatmap(data_for_heatmap, cluster_rows = TRUE, cluster_cols = TRUE, show_rownames = FALSE, color = my_palette)

# Save the heatmap as a PNG file with transparent background
png("heatmap.png", width = 800, height = 600, bg = "transparent") 
print(heatmap) 
dev.off()  
```
